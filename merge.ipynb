**Inner Join or Natural join: To keep only rows that match from the data frames, specify the argument how=‘inner’, on= specify the column the dataframes have in common. **

```python
# inner join in python pandas which is the default merge (joining the order and product dataframe)
df1 = pd.merge(Orders, products, on='product_id', how='inner')
df1
```

```python
# Alternativerly, we could use this syntax for inner join since its the default join (joining the order and product dataframe)
df2 = pd.merge(Orders, products)
df2
```

From the result above, we discovered that only items the both dataframe had in common was returned

```python
# joining the resulting dataframe from the order and product dataframe join with the product dataframe using inner join 
df4 = pd.merge(df1, customers)
df4
```

```python
# Alterntevely, we could join the resulting dataframe from the order and product dataframe join with the product dataframe using inner join using the syntax below
df3 = pd.merge(df1, customers, on='customer_id', how='inner')
df3
```

```python
# Perform an inner merge operation between customers and orders tables. Here you need to specify the columns you’ll be merging on explicitly 

df5 = pd.merge(Orders, customers, on='customer_id', how='inner')
df5
```

From the result above, we discovered that only items the both dataframe have in common was returned using inner join


```python
# Perform an outer merge between customers and products tables.
df_merg = pd.merge(customers, products, how='outer')
print(df_merg)
```

```python
# Perform a left merge between orders and customers tables.
df6 = pd.merge(Orders, customers, on='customer_id', how='left')
df6
```

All rows from the left table which is the order table, and any rows with matching keys from the right table which is the customers table.When there is no Matching from right table NaN will be returned

```python
# Perform a right merge between product and customers tables.
df6 = pd.merge(products, customers, how='left')
print(df6)
```

```python
# We can join the product and customer table using join based on index
 
df_index = pd.merge(products, customers, left_index=True, right_index=True )
df_index
```
